### 2) Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release) and  make data frame. 

```python
def moive(url):
    #send get request to the webpage server to get the source code of the page.
    page=requests.get(url)

    #page content
    soup=BeautifulSoup(page.content)

    #scrap movie names of Top rated indian movies in IMDB
    names=[]
    for i in soup.find_all('td',class_="titleColumn"):
        names.append(i.text.split()[1:-1])
    for i in range(0,len(names)):
        names[i]=' '.join(names[i])
    
    #scrap movie rating of Top rated indian movies in IMDB
    rating=[]
    for i in soup.find_all('td',class_="ratingColumn imdbRating"):
        rating.append(i.text.split()[0])

    #converting string to float
    for i in range(0,len(rating)):
        rating[i]=float(rating[i])
    #print(rating)

    #scrap year of release
    year=[]
    for i in soup.find_all('td',class_="titleColumn"):
        year.append(i.text.replace('(','').replace(')','').split()[-1])

    #converting string to integer
    for i in range(0,len(year)):
        year[i]=int(year[i])
    #print(year)

    #checking length of all columns
    #print(len(names),len(rating),len(year))

    #import pandas library
    import pandas as pd
    df=pd.DataFrame({'Movie Names':names,'Movie Ratings':rating,'Year of Release':year})
    df=df.head(100)

    #printing top rated 100 indian movies
    return df
```

```python
print("Top 100 rated movies in IMDB's")
moive('https://www.imdb.com/chart/top/')
```

### 4) Write a python program to scrape product name, price and discounts from  https://meesho.com/bagsladies/pl/p7vbp .

### 5) Write a python program to scrape cricket rankings from icc-cricket.com. You have to scrape: 

#### a) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating. 

```python
def team_url(url):
    #send get request to the webpage server to get the source code of the page.
    page=requests.get(url)

    #page content
    soup=BeautifulSoup(page.content)

    #scrape Team Name
    team=[]
    for x in soup.find_all('span',class_="u-hide-phablet"):
        team.append(x.text)


    #scrape number of matches
    match=[]
    m1=soup.find('td',class_="rankings-block__banner--matches")
    match.append(m1.text)

    for i in soup.find_all('tr',class_="table-body"):
        match.append(i.text.split()[-3])

     #converting string to integer
    for i in range(0,len(match)):
        match[i]=int(match[i])
    #print(match)

    #scrape number of points
    points=[]
    p1=soup.find('td',class_="rankings-block__banner--points")
    points.append(p1.text)

    for i in soup.find_all('tr',class_="table-body"):
        points.append(i.text.split()[-2])
    
    #scrape Team rating
    rating=[]
    r1=soup.find('td',class_="rankings-block__banner--rating u-text-right")
    rating.append(r1.text)

    for i in soup.find_all('tr',class_="table-body"):
        rating.append(i.text.split()[-1])

    #converting string to float
    for i in range(0,len(rating)):
        rating[i]=float(rating[i])
    #print(rating)

    #To remove extra rows in team column,we use pop operation
    for x in range(20,24):
        team.pop()
    team.pop(-1)

    #checking len before making dataframe
    #print(len(team),len(match),len(points),len(rating))


    #import pandas
    import pandas as pd
    df=pd.DataFrame({'Teams':team,'Number of Matches':match,'Team Points':points,'Team Rating':rating})

    #for selecting Top 10 teams,we use Head()
    df=df.head(10)
    return df
```

```python
print('Top 10 ODI teams in men’s cricket along with the records for matches, points and rating')
team_url('https://www.icc-cricket.com/rankings/mens/team-rankings/odi')
```

#### b)Top 10 ODI Batsmen along with the records of their team and rating

```python
def cric_url(url):
    #send get request to the webpage server to get the source code of the page.
    page=requests.get(url)

    #page content
    soup=BeautifulSoup(page.content)

    #player name
    name=[]
    n1=soup.find('div',class_="rankings-block__banner--name-large")
    name.append(n1.text)

    for x in soup.find_all('td',class_="table-body__cell rankings-table__name name"):
        name.append(x.text.replace('\n',''))

    #scrape player team name
    team=[]
    t1=soup.find('div',class_="rankings-block__banner--nationality")
    team.append(t1.text.split()[0])

    for x in soup.find_all('td',class_="table-body__cell nationality-logo rankings-table__team"):
        team.append(x.text.replace('\n',''))

    #scrape player rating
    rating=[]
    r1=soup.find('td',class_="u-text-left")
    rating.append(r1.text.replace('\n',''))

    for x in soup.find_all('td',class_="table-body__cell rating"):
        rating.append(x.text)

    #convert string to integer
    for i in range(0,len(rating)):
        rating[i]=int(rating[i])

    #check all column length
    #print(len(name),len(team),len(rating))

    #import pandas
    import pandas as pd

    #make dataframe
    df=pd.DataFrame({'Player Name':name,'Player Team':team,'Player Rating':rating})

    #Top 10 ODI players along with the records of their team and rating
    df=df.head(10)
    return df
```

```python
print("Top 10 ODI Batsmen along with the records of their team and rating")
cric_url('https://www.icc-cricket.com/rankings/mens/player-rankings/odi/batting')
```

### 7) Write a python program to scrape details of all the posts from coreyms.com. Scrape the heading, date, content and the code for the video from the link for the youtube video from the post. 

```python
#send get request to the webpage server to get the source code of the page.
page=requests.get('https://coreyms.com/')

#page content
soup=BeautifulSoup(page.content)

#scrape posts headings
heading=[]
for i in soup.find_all('h2',class_="entry-title"):
    heading.append(i.text)
#remove one heading bcz it is not post
heading.pop(4)

#scrape posts date
date=[]
for i in soup.find_all('time',class_="entry-time"):
    date.append(i.text)

#remove one date bcz it is not post
date.pop(4)

#scrape posts content
content=[]
for i in soup.find_all('div',class_="entry-content"):
    content.append(i.text.replace('\n',''))

#remove one content bcz it is not post
content.pop(4)

#scrape the post link for youtube video
link=[]
for i in soup.find_all('iframe',class_="youtube-player"):
    link.append(i.get('src'))

#checking length of all columns
#print(len(heading),len(date),len(content),len(link))

#import pandas
import pandas as pd

#make dataframe
df=pd.DataFrame({'Poster Heading':heading,'Posted Date':date,'Poster Content':content,'Poster link':link})

# The heading, date, content and the code for the video from the link for the youtube video from the post
print('The heading, date, content and the code for the video from the link for the youtube video from the post')
df
```

### 8) Write a python program to scrape house details from mentioned URL. It should include house title, location, area, EMI and price from https://www.nobroker.in/ .Enter three localities which are Indira Nagar, Jayanagar, Rajaji Nagar. 

```python
def locality(url):
    #send get request to the webpage server to get the source code of the page.
    page=requests.get(url)
    
    #page content
    soup=BeautifulSoup(page.content)
    
    #scrape house title
    title=[]
    for x in soup.find_all('span',class_="overflow-hidden overflow-ellipsis whitespace-nowrap max-w-80pe po:max-w-full"):
        title.append(x.text)
    #scrape house location
    location=[]
    for x in soup.find_all('div',class_="mt-0.5p overflow-hidden overflow-ellipsis whitespace-nowrap max-w-70 text-gray-light leading-4 po:mb-0 po:max-w-95"):
        location.append(x.text.replace('\xa0',''))
    #scrape house area
    area=[]
    for x in soup.find_all('div',class_="font-semi-bold heading-6"):
        area.append(x.text)
    area=area[::3]
    #scrape house EMI
    emi=[]
    for x in soup.find_all('div',class_="font-semi-bold heading-6"):
        emi.append(x.text)
    emi.pop(0)
    emi=emi[::3]
    #scrape house price
    price=[]
    for x in soup.find_all('div',class_="font-semi-bold heading-6"):
        price.append(x.text)
    price.pop(0)
    price.pop(0)
    price=price[::3]
    
    #checking length of all columns
    #print(len(title),len(location),len(area),len(emi),len(price))
    
    #import pandas
    import pandas as pd
    
    #make dataframe
    df=pd.DataFrame({'House Title':title,'House Location':location,'House Area':area,'House EMI':emi,'House Price':price})
    return df
```

### 9) Write a python program to scrape mentioned details from dineout.co.in : 

```python
#send get request to the webpage server to get the source code of the page.
page=requests.get('https://www.dineout.co.in/delhi-restaurants/welcome-back')

#page content
soup=BeautifulSoup(page.content)

#scrape Restaurant name
name=[]
for x in soup.find_all('a',class_="restnt-name ellipsis"):
    name.append(x.text)

#scrape Cuisine
cuisine=[]
for x in soup.find_all('span',class_="double-line-ellipsis"):
    cuisine.append(x.text.split()[6::])
for i in range(0,len(cuisine)):
    cuisine[i]=' '.join(cuisine[i])

#scrape restaurant location
location=[]
for x in soup.find_all('div',class_="restnt-loc ellipsis"):
    location.append(x.text)

#scrape restaurant rating
rating=[]
for x in soup.find_all('div',class_="restnt-rating rating-4"):
    rating.append(x.text)

#change string to float
for i in range(0,len(rating)):
    rating[i]=float(rating[i])

#scrape restaurant image link
image=[]
for i in soup.find_all('img',class_="no-img"):
    image.append(i.get('data-src'))

#check all columns length before making dataframe
#print(len(name),len(cuisine),len(location),len(rating),len(image))

#import pandas
import pandas as pd

#make dataframe
df=pd.DataFrame({'Restaurant Name':name,'Restaurant Cuisine':cuisine,'Restaurant Location':location,'Restaurant Rating':rating,'Restaurant Image link':image})
print('Restaurant Name,Cuisine,Location,Rating,Image link ')
df
```

### 10) Write a python program to scrape first 10 product details which include product name , price , Image URL from https://www.bewakoof.com/women-tshirts?ga_q=tshirts .  

```python
#send get request to the webpage server to get the source code of the page.
page=requests.get('https://www.bewakoof.com/women-clothing')

#page content
soup=BeautifulSoup(page.content)

#scrape T-shirt name
name=[]
for x in soup.find_all('div',class_="productCardDetail"):
    name.append(x.text.replace('₹','').split()[0:-4])
for i in range(0,len(name)):
    name[i]=' '.join(name[i])

#scrape T-shirt price
price=[]
for x in soup.find_all('span',class_="discountedPriceText"):
    price.append(x.text.replace('₹ ',''))
#scrape T-shirt image URL
image=[]
for x in soup.find_all('img',class_="productImgTag"):
    image.append(x.get('src'))

#check the lengths before making dataframe
#print(len(name),len(price),len(image))

#import pandas
import pandas as pd

#make dataframe
df=pd.DataFrame({'T-shirt Name':name,'T-shirt price':price,'T-shirt image URL':image})
print('First 10 product details which include product name , price , Image URL')
df
```
